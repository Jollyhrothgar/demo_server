```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from random import sample
```

```python
# Table-like output
d1 = {
    'a':[1,2,3],
    'b':[4,5,6],
    'c':[7,8,9]
}

# 1D data: histogram or maybe even plot?
base = np.arange(0,10)

# Plot OR Table like Output
d2 = {
    'vals':base,
    'more':base**2
}

# plottable output
tup = (base,np.sin(base))

# mappable output
def generate_gps_list(num=10,lat_var=90./10000., lon_var=180./10000. , center=None):
    """
    Generate gaussian distribution of GPS coordinates around a central
    point. If no central point is supplied, 
    
    param center: tuple (lat,long)
    param lat_var: variance of points generated in lattitude
    param long_va: varience of points generated in longitude
    """
    sign = np.array([-1,1])
    
    if center is None:
        center = (
            np.random.choice(np.linspace(start=-90, stop=90, num=10000)),
            np.random.choice(np.linspace(start=-180,stop=180,num=10000))
        )
    elif(len(center) is not 2):
        raise ValueError("center must be a tuple of length 2")
    list_of_list_pairs = []
    list_of_tuples = []
    lat_list = []
    lon_list = []
        
    for i in range(num):
        lat, lon = (
                center[0]+np.random.normal(loc=center[0], scale=lat_var**(0.5)),
                center[1]+np.random.normal(loc=center[1], scale=lon_var**(0.5))
            )
        list_of_list_pairs.append([lat,lon])
        list_of_tuples.append((lat,lon))
        lat_list.append(lat)
        lon_list.append(lon)
    return list_of_list_pairs, list_of_tuples, lat_list, lon_list
```
